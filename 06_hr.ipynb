# HR-аналитика для компании работа с заботой

Нам передали данные о сотрудниках в нескольких таблицах. Основные данные о сотруднкиах:

- id — уникальный идентификатор сотрудника;
- dept — отдел, в котором работает сотрудник;
- level — уровень занимаемой должности;
- workload — уровень загруженности сотрудника;
- employment_years — длительность работы в компании (в годах);
- last_year_promo — показывает, было ли повышение за последний год;
- last_year_violations — показывает, нарушал ли сотрудник трудовой договор за последний год;
- supervisor_evaluation — оценка качества работы сотрудника, которую дал руководитель;
- salary — ежемесячная зарплата сотрудника;
- job_satisfaction_rate — уровень удовлетворённости сотрудника работой в компании, целевой признак.
- quit — увольнение сотрудника из компании, целевой признак во второй задаче.



***Цель исследования***:
1. Проанализировать полученные данные
2. Оценить удовлетворенность сотрудника работой на основе предоставленных данных, создав для этого модель МО
3. Составить портрет уволившегося сотрудника
4. Построить модель для нахождения сотрудников, которые скорее всего уйдут

***Ход исследования***:
1. Получим данные из исходных датасетов
2. Предобработаем данные: поищем пропуски, дубликаты.
3. Проведем исследовательский анализ данных всех полученных датасетов
4. Поищем корреляцию между признаками чтобы не стоить модель по походим признакам
5. Подготовим данные для построения модели МО с использованием пайплайнов, рассмотрим несколько моделей и параметров
6. Оценим полученные модели с помощью метода SMAPE (для первой) и ROC_AUC (для второй)
7. Сделаем общий вывод по работе, дадим рекомендации сервису по удержанию сотрудников

##### Функции для построения графиков

```python
def figures_for_num(data, numeric_col):
    if len(numeric_col) != 1:
        fig, axes = plt.subplots(len(numeric_col), 2, figsize=(10,len(numeric_col)*5))
        i = 0
        plt.subplots_adjust(hspace=0.3)
        for column in numeric_col:
            axes[i,0].set_title(f'KDE для параметра \n {column}')
            axes[i,1].set_title(f'Boxplot для параметра \n {column}')
            sns.kdeplot(ax = axes[i,0], data=data, x=column)
            sns.boxplot(ax = axes[i,1], data=data, x=column)
            i += 1
    else:
        column = numeric_col[0]
        fig, axes = plt.subplots(1, 2, figsize=(10,4))
        plt.subplots_adjust(hspace=0.3)
        axes[0].set_title(f'KDE для параметра \n {column}')
        axes[1].set_title(f'Boxplot для параметра \n {column}')
        sns.kdeplot(ax = axes[0], data=data, x=column)
        sns.boxplot(ax = axes[1], data=data, x=column)
```

```python
def figures_for_category(data, category_col):
    if len(category_col) != 1:
        fig, axes = plt.subplots(len(category_col), figsize=(7,len(category_col)*5))
        plt.subplots_adjust(hspace=0.3)
        i = 0
        for column in category_col:
            axes[i].set_title(f'Гистограмма для параметра \n {category_col[i]}')
            ax = sns.countplot(ax = axes[i], data=data, x=column,
                               order=data[column].value_counts(ascending=False).index)
            abs_values = data[column].value_counts(ascending=False).values
            ax.bar_label(container=ax.containers[0], labels=abs_values)
            i += 1
    else:
        column = category_col[0]
        ax = sns.countplot(data=data, x=column,
                           order=data[column].value_counts(ascending=False).index)
        abs_values = data[column].value_counts(ascending=False).values
        ax.bar_label(container=ax.containers[0], labels=abs_values)  
        ax.set_title(f'Гистограмма для параметра \n {column}')
```

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
) 
```

```python
final_pipe = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeRegressor())
    ]
) 
```

```python
def smape_metric_score(y_true, y_pred):
    return 100/len(y_true) * np.sum(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))
```

```python
smape_scorer = make_scorer(smape_metric_score, response_method='predict', greater_is_better=False)
```

```python
param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 15),
        'models__min_samples_leaf': range(2, 5),
        'models__min_samples_split': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # словарь для модели SVR()
    {
        'models': [SVR()],
        'models__C': [0.01, 0.1, 1.0, 10],
        'models__kernel': ['rbf', 'sigmoid'],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  

    },
    # словарь для модели poly SVR()
    {
        'models': [SVR(kernel='poly')],
        'models__C': [0.01, 0.1, 1.0, 10],
        'models__degree': range(2,4),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  

    },
    # словарь для модели linear SVR() очень долго считает с passthrough, поэтому убрал
    {
        'models': [SVR(kernel='linear')],
        'models__C': [0.01, 0.1, 1.0, 10],
        'preprocessor__num': [StandardScaler(), MinMaxScaler()]  

    },
    # словарь для модели LinearRegression()
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
```

Далее был построен пайплайн и применен поиск по сетке для выявления лучшей модели МО для прдсказания удовлетворенности работой. В результате построения моделей МО была получена лучшая модель по метрике SMAPE(symmetric mean absolute percentage error): DecisionTreeRegressor(max_depth=14, min_samples_leaf=3, random_state=42). Метрика на тестовых данных составила 13.7, а при кросс-валидации 14.3. В результате работы данной модели предсказали удовлетворенность сотрудника своей работой для дальнейшего использования. Данная модель справилась лучше, так как в этих данных зависимости лучше описываются деревом решений, чем линейной регрессии, например из-за нелинейных связей между данными.

Затем были аналогично проанализированы данные второй тренировочной таблицы. Распределение всех параметров очень слабо отличалось от второй. К ней присоединили данные из предсказания удовлетворенности, построили график зависимости от нее для сотрудников ушедших из компании и оставшихся в компании. На основании графика подтверждено предположение о том, что удовлетворенность влияет на уход сотрудника. Типичный уволившийся сотрудник - джун, работающий менее года без повышений с нарушениями договора с низкой загруженностью и плохой оценкой супервизора.

После этого был построен второй пайплайн для предсказания ухода. В нем также выбиралась лучшая модель. Ей стала RandomForestClassifier(max_depth=6, min_samples_leaf=2, random_state=42) со значение метрики ROC_AUC 0.929 на тестовых данных. Параметры, которые использовались для предсказания, были частично переделаны в полиномы второй степени. По анализу результатов с помощью shap получено, что основные признаки это: job_satisfaction_rate, level, workload, employment_years, salary. Также стоит отметить признак повышений: большие значения этого признака склоняют модель к исходу 0 (сотрудник не уволиться), а низкие практически не увеличивают вероятность ухода по сравнению с остальными параметрами.

Таким образом, для удержания сотрудников рекомендуется поднимать их уровень удовлетворенности работой, а также следить, чтобы они не были недозагружены работой. В первые три года желательно проводить повышения, так как в этот период увольняются чаще всего, а наличие повышения сильно снижает вероятность ухода.
