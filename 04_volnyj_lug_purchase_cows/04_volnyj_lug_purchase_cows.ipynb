# Разработка модели МО для отбора коров

Нам передали данные о коровах на ферме "Вольный луг" и пастбище "ЭкоФерма". Необходимо построить модель МО для отбора коров на пастбище, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, а её надой был вкусным — строго по его критериям, ничуть не хуже. 
Нам предоставлены следующие данные о коровах:
- Характеристики коровы: id, Порода, порода папы_быка, Возраст.
- Характеристики корма: ЭКЕ (Энергетическая кормовая единица), Сырой протеин, г, СПО (Сахаро-протеиновое соотношение).
- Характеристика пастбища: Тип пастбища.
- Характеристики молока: Удой, кг, Жирность,%, Белок,%, Вкус молока.

***Цель исследования***:
1. Проанализировать полученные данные
2. Провести корреляционный анализ датасета с кровами на ферме
3. Построить модели линейной регрессии для удоя
4. Построить модель классификации для вкуса молока

***Ход исследования***:
1. Получим данные из исходных датасетов о коровах на ферме ('datasets/ferma_main.csv'), папах коров ('datasets/ferma_dad.csv') и коровах на пастбище ('datasets/cow_buy.csv')
2. Обработаем исходные данные: поищем в них ошибки, пропуски и дубликаты, в случае необходимости исправим.
3. Проведем исследование всех признаков в датасете с коровами с фермы, построим для них графики. Поищем выбросы, необчные распределения данных, попробуем их улучшить
4. Сделаем корреляционный анализ всех данных, поищем линейную связь, мультиколлинеарность
5. Обучим 3 модели регрессии с разными параметрами. Первую по имеющимся данным, вторую с исключением нелинейности для признаков СПО и ЭКЕ, в третью добавим новый признак имя папы
6. Сравним качество моделей и выберем лучшую, на ее основе предскажем удой
7. Создадим модель классификации, предскажим вкус молока, также попытаемся минимизировать количество ложных определений вкусного молока
8. Сделаем общий вывод, дадим рекомендации по закупке

```python
def train_model(data, quantitative_cells, category_cells):
    X = data.drop(columns=['milk_yield_kg', 'fat_content_%', 'protein_%'], axis=1)
    y = data['milk_yield_kg']
    model_lr  = LinearRegression()
    X_train, X_test, y_train, y_test = train_test_split(
        X, 
        y, 
        random_state=RANDOM_STATE
    )
    
    X_train_scaled = scaler.fit_transform(X_train[quantitative_cells])
    X_test_scaled = scaler.transform(X_test[quantitative_cells])
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=quantitative_cells)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=quantitative_cells)

    X_train_ohe = encoder.fit_transform(X_train[category_cells])
    X_test_ohe = encoder.transform(X_test[category_cells])
    encoder_col_names = encoder.get_feature_names()
    X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
    X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

    X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
    X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)

    print(encoder_col_names)
    
    return model_lr, X_train, X_test, y_train, y_test, encoder_col_names
```

```python
model_lr_1, X_train, X_test, y_train, y_test, encoder_col_names = (
    train_model(ferma_main, quantitative_cells, category_cells)
)
```

```python
model_lr_1.fit(X_train, y_train)
predictions_1 = model_lr_1.predict(X_test)
residuals_1 = y_test - predictions_1
```

```python
r2_1 = r2_score(y_test, predictions_1)
mae_1 = mean_absolute_error(y_test, predictions_1)
rmse_1 = mean_squared_error(y_test, predictions_1, squared=False)
mse_1 = mean_squared_error(y_test, predictions_1)
print(f'Значение R2 = {r2_1}')
print(f'Значение MAE = {mae_1}')
print(f'Значение RMSE = {rmse_1}')
print(f'Значение MSE = {mse_1}')
```

```python
def residuals_graphics(residuals, predictions, y_test):
    residuals = y_test - predictions

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
    axes[0].hist(residuals)
    axes[0].set_title('Гистограмма распределения остатков')
    axes[0].set_xlabel('Остатки')

    axes[1].scatter(predictions, residuals)
    axes[1].set_xlabel('Предсказания модели')
    axes[1].set_ylabel('Остатки')
    axes[1].set_title('Анализ дисперсии')
    plt.show()
residuals_graphics(residuals_1, predictions_1, y_test)
```

Для первой модели получение значение $R^2$ близкое к еденицы. По графикам можем видеть нормальное распределение остатков около 0, но с небольшим смещением влево.

```python
confusion = np.zeros((2, 2)).astype(np.int64)
for i, k in enumerate([0, 1]):
    for j, l in enumerate([0, 1]):
        confusion[i][j] = np.sum(np.logical_and(y_test == k, y_pred == l))
print(confusion)
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.rcParams['axes.grid'] = False
cmd = ConfusionMatrixDisplay(cm)
ax = cmd.plot()
plt.rcParams['axes.grid'] = True
```

Так как нам необходимо, чтобы молоко обязательно было вкусным, необходимо минимизировать ложные положительные результаты, то есть улучшить метрику precision.

```python
y_proba = clf.predict_proba(X_test)[:,1] 
data = pd.DataFrame(zip(y_test, y_pred, y_proba),
             columns = ['y_valid', 'y_pred', 'y_proba']).sort_values(by='y_proba',ascending=False)
data.head(20)
```

Видим первую возникшую ошибку при вероятности 0.804, поэтому возьмем порог для верных предсказаний 0.81. Дополнительно выведем результаты с разными границами, чтобы на примере показать, почему для данного случае берем 0.81

```python
thresholds = [round(i,2) for i in np.linspace(0.6,1,num = 5,endpoint=False)]
columns = []
for i in thresholds:
    columns.append('y_pred_'+str(i))
    data['y_pred_'+str(i)] = data['y_proba'].apply(lambda x: 1 if x >= i else 0)
data.head(20)
```

```python
values = []
thresholds = list(data['y_proba'])
thresholds.sort()
for i in thresholds:
    data['y_pred_'] = data['y_proba'].apply(lambda x: 1 if x>= i else 0)
    values.append(precision_score(data['y_valid'], data['y_pred_']))
heatmap = sns.lineplot(x=thresholds, y=values)
plt.ylabel('Precision')
plt.xlabel('Порог')
plt.title('Зависимость precision от выбранного порога')
```

На графике можем наблюдать постепенный рост метрики precision с небольшими падениями, которые вызваны уменьшением общего количества положительных результатов модели. Также можем видеть при пороге 0.81 выход на прямую: тут значение precision уже равно единицы, так как все значения предсказаны верно.

```python
data['y_pred_0.81'] = data['y_proba'].apply(lambda x: 1 if x>= 0.81 else 0)

recall = recall_score(data['y_valid'], data['y_pred_0.81'])
precision = precision_score(data['y_valid'], data['y_pred_0.81'])

print(f'recall: {round(recall ,3)}')
print(f'precision: {round(precision,3)}')
```

Полученный recall 0.091 говорит о наличии большого количества ошибок второго рода. Precision же в данном случае 1, то есть ошибок первого рода нет.
